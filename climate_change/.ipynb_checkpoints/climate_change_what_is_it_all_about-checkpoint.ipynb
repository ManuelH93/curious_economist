# Climate change - what is it all about

## Import libaries

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

## Import data

```python
temperature = pd.read_csv('raw_data/Complete_TAVG_complete.csv', sep=',')
```

```python
temperature.head()
```

## Plot temperature data

```python
# Graph version 2
plt.figure(figsize=(60,30))
#plt.xlabel('Liquidity')


plt.subplot(1, 1, 1)
ax2 = temperature["monthly_anomaly"].plot(color=((56/255),(177/255),(186/255)), grid=True, label='temperature anomaly',
                                         linewidth=4)
ax2.tick_params(axis='both', which='major', labelsize=12)
h2, l2 = ax2.get_legend_handles_labels()
#ax2.set_ylim(0, 0.4)
plt.legend(h2, l2, loc=2)
plt.xticks(rotation=20)
plt.xlabel('month')

#plt.savefig('liquidity.png', bbox_inches='tight')
plt.show()
```
